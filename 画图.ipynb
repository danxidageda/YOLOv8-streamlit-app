```python
import matplotlib.pyplot as plt
from PIL import Image

def show_image_comparison(image_paths, titles, rows, cols, figsize=(14, 10)):
    num_images = len(image_paths)

    if num_images != len(titles):
        raise ValueError("Number of images must match number of titles")

    if num_images > rows * cols:
        raise ValueError("Too many images for the given number of rows and columns")

    fig, axes = plt.subplots(rows, cols, figsize=figsize)

    for i, (image_path, title) in enumerate(zip(image_paths, titles)):
        row = i // cols
        col = i % cols
        image = Image.open(image_path)
        axes[row, col].imshow(image)
        axes[row, col].axis('off')
        axes[row, col].set_title(title)

    plt.tight_layout()
    plt.show()

# 输入多张照片的路径和对应的标题
image_paths = [
    r'E:\code_for_school\leaf_detection\ultralytics-main\datasets\15002000data\train\images\data002_0_0_2000_1500.png',
    r'E:\code_for_school\leaf_detection\ultralytics-main\datasets\15002000data\train\images\data002_0_0_2000_1500_blurred.jpg',
    r'E:\code_for_school\leaf_detection\ultralytics-main\datasets\15002000data\train\images\data002_0_0_2000_1500_motion_blur.jpg',
    r"E:\code_for_school\leaf_detection\ultralytics-main\datasets\15002000data\train\images\data002_2000_0_4000_1500.png"
    # 添加更多图片路径...
]

titles = [
    'Title for Image 1',
    'Title for Image 2',
    'Title for Image 3',
    'Title for Image 3'
    # 添加更多图片标题...
]

# 设置每行和每列显示的图片数量
rows = 2
cols = 2

# 显示对比图
show_image_comparison(image_paths, titles, rows, cols)
```
